```python
import numpy as np
def diffusion_kalman_filter(psi, x_hat, P, N, H, R, y, A, Q, W, num_steps):
  num_nodes = len(x_hat)
  prediction_hist = np.zeros((num_steps,len(x_hat),x_hat[0].shape[0],x_hat[0].shape[1]))
  covariance_hist = np.zeros((num_steps,len(P),P[0].shape[0],P[0].shape[1]))
  # set loop for all nodes i
  for k in range(num_steps):
      # Measurement Sharing and Update Step
      for i in range(num_nodes):
        psi_i = psi[i]
        P_i = P[i]

        for j in N[i]:
          # Compute Kalman gain
          Sij = (H[j] @ P_i @ H[j].T) + R[j]

          Kij = P_i @ H[j].T @ np.linalg.inv(Sij)
          # Update state estimate
          t_j = y[k][j] - (H[j] @ psi_i)  
          psi_i = psi_i + (Kij @ t_j)

          # Update error covariance
          P_i = (np.eye(P_i.shape[0]) - (Kij @ H[j])) @ P_i

          print(Kij, (P_i @ H[j].T @ np.linalg.inv(R[j])))

          psi[i] = psi_i
          P[i] = P_i

      # Estimate Sharing Step
      new_x_hat = [0] * num_nodes
      for i in range(num_nodes):
        new_x_hat[i] = np.sum([W[i, j] * psi[j] for j in N[i]], axis=0)

        x_hat = new_x_hat

      # Prediction Step
      for i in range(num_nodes):
        psi[i] = A @ x_hat[i]
        prediction_hist[k,i] = psi[i]
        P[i] = (A @ P[i] @ A.T) + Q
        covariance_hist[k,i] = P[i]

  return prediction_hist, covariance_hist
```
